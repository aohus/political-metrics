```python

import json
import os
import pandas as pd

bills = pd.read_json("./data/assembly/formatted/bills.json", orient="records")
bill_details = pd.read_json("./data/assembly/formatted/bill_details.json", orient="records")
proposers = pd.read_json("./data/assembly/formatted/proposer_bills.json", orient="records")
members = pd.read_json("./data/assembly/raw/cur_members.json", orient="records")
```

```python
with open("./data/law/raw/cur_law.json", "r", encoding="utf-8") as f:
    data = json.load(f)
law = []
for row in data:
    law.extend(row["LawSearch"]["law"])
```

```python
법률, 시행령, 시행규칙 = [], [], []
for item in law:
    law_type = item["법령구분명"]
    if law_type == "법률":
        법률.append(item)
    elif law_type.endswith("령"):
        시행령.append(item)
    elif law_type.endswith("규칙"):
        시행규칙.append(item)
    else:
        print(item)
len(법률), len(시행령), len(시행규칙)
```

```python
multiple = []
comms = dict()
for row in 법률:
    codes = row["소관부처코드"].split(",")
    names = row["소관부처명"].split(",")
    if len(codes) > 1:
        multiple.append(tuple(names))
    else:
        if codes[0] not in comms:
            comms[codes[0]] = names[0]
        else:
            if comms[codes[0]] != names[0]:
                print(f"Conflict: {codes[0]} - {comms[codes[0]]} vs {names[0]}")    
```

```python
cur_law = {}
for item in 법률:
    cur_law[item["법령명한글"]] = item
```

```python
from concurrent.futures import ThreadPoolExecutor
import os
import json

input_dir = "./data/document/parsed"
output_dir = "./data/document/text"

def save_full_text(fname):
    with open(os.path.join(input_dir, fname), "r") as f:
        data = json.load(f)
    full_text = data["full_text"]
    full_text = full_text.replace("\n\n\n\n", " ").strip()
    with open(os.path.join(output_dir, fname.split(".")[0]+".txt"), "w", encoding="utf-8") as f:
        f.write(full_text)

with ThreadPoolExecutor() as executor:
    executor.map(save_full_text, os.listdir(input_dir))
```
