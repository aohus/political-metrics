```python

import json
import os
import pandas as pd
```

```python
excel_path = "./data/assembly/raw/activity.xlsx"
df = pd.read_excel(excel_path, sheet_name="의원실 행사 정보")
df.columns = [col.strip() for col in df.columns]  # Clean column names
```

```python
df.rename(columns={
    "국회의원명": "NAAS_NM",
    "행사 제목": "EV_TTL",
    "행사 일시": "EV_DTM",
    "행사 장소": "EV_PLC",
    "링크URL": "LINK_URL"}, inplace=True)
```

```python
df["EV_DT"] = df["EV_DTM"].str[:10]
df["EV_DT"] = df["EV_DT"].str.replace(".", "-")
df["EV_DT"] = pd.to_datetime(df["EV_DT"], errors='coerce')
df["EV_DM"] = df["EV_DTM"].str[12:-1]
```

```python
df.to_json("./data/assembly/formatted/activity.json", orient="records", force_ascii=False)
```
